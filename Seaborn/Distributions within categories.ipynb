## Distributions within categories:

- We've explored distribution plots for a single feature, but what if we want to compare distributions across categories
- For example, instead of the distribution of everyone's salary, we can compare the distributions of salary per level of education.
- We will first separate out each category, then create the distribution visualization.


### Types of plots:

1. Boxplot
2. Violinplot
3. Swarmplot
4. Boxenplot (Letter-Value plot)



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/home/legacy/anaconda3/Data Science Master Class/UNZIP_FOR_NOTEBOOKS_FINAL/05-Seaborn/StudentsPerformance.csv')
```

### Boxplots

- The boxplot displays the distribution of a continuous variable.
- It does this through the use of quartiles.
- Quartiles separate the data into 4 equal  number of data points:
  - 25% of data points are in the bottom quartile.
  - 50th percentile (Q2) is the median.
- We can create a boxplot per category.

### Violinplots:
- The violinplot plays a similar role as the box plot.
- It displays the probability density across the data using a KDE.
- We can imagine it as a mirrored KDE plot.
- It looks interesting. Shows KDE + boxplot in the middle.

### Swarmplots:
- It is very simply, it shows all the data points in the distribution.
-  For very large datasets, it won't show all the points, but will display the general distribution of them.
- Looks like a violinplot, but with points.
- But take into account that it is not a mirrored image, however, it looks like it is.
- It allows you to know exactly the amount of dots you have.


### Boxenplots:
- It was recently devleoped (2011)
- There is a paper linked in the notebook if you need more information about this plot.


Keep in mind that the main goal of data visualization is to inform, not to confuse or show-off various esoteric plots!

```python
# Simple boxplot:
plt.figure(figsize=(12,6), dpi=200)
sns.boxplot(data=df, y='math score', x='test preparation course')
```

```python
# Or you could also do:
plt.figure(figsize=(12,6), dpi=200)
sns.boxplot(data=df, y='math score', x='parental level of education')
```

```python
# Or you could also include another category by adding the hue argument:
plt.figure(figsize=(12,6), dpi=200)
sns.boxplot(data=df, y='math score', x='parental level of education', hue='test preparation course')
```

```python
# What if we swap x and y axis:
plt.figure(figsize=(12,6), dpi=200)
sns.boxplot(data=df, x='math score', y='parental level of education', hue='test preparation course')
```

#### Let's now code a violin plot:

```python

plt.figure(figsize=(12,6), dpi=200)
sns.violinplot(data=df, y='math score', x='parental level of education')
```

```python
# With hue:
plt.figure(figsize=(12,6), dpi=200)
sns.violinplot(data=df, y='math score', x='parental level of education', hue='test preparation course')
```

```python
# If you split it, you will have only one assymetrical violin including the category and hue (sub-category):
plt.figure(figsize=(12,6), dpi=200)
sns.violinplot(data=df, y='math score', x='parental level of education', hue='test preparation course', split=True)
```

```python
# You can add quartiles:
# If you split it, you will have only one assymetrical violin including the category and hue (sub-category):
plt.figure(figsize=(12,6), dpi=200)
sns.violinplot(data=df, y='math score', x='parental level of education', hue='test preparation course'
               , split=True, inner='quartile')
```

```python
#### Now, let's code a swarmplot:
plt.figure(figsize=(12,6), dpi=200)
sns.swarmplot(data=df, x='math score', y='gender', size=2, hue='test preparation course') #It is not easy to read
```

```python
# To make the previous plot easier to read, use dodge=True argument:
plt.figure(figsize=(12,6), dpi=200)
sns.swarmplot(data=df, x='math score', y='gender', size=2, hue='test preparation course', dodge=True) #It is not easy to read
```

#### Boxenplot

```python
sns.boxenplot(data=df, x='math score', y='test preparation course')
```
