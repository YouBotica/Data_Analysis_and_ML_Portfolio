## Categorical Plots
### Statistical estimation within categories

- The categorical plots are going to focus on displaying statistical metrics per category.
- For example: mean value per category, or a count of the number of rows per category.
- It is the equivalent of a groupby() call

The two main types of plots for this are:
1. countplot()
  - Counts number of rows per category.

2. barplot()
  - General form of displaying any chosen metric per category.
  
You can further add hue to countplots and barplots to have an extra category, and thus, give more information.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/home/legacy/anaconda3/Data Science Master Class/UNZIP_FOR_NOTEBOOKS_FINAL/05-Seaborn/dm_office_sales.csv')
```

```python
df.head()
```

```python
# In Pandas:
df['division'].value_counts()
```

```python
# In seaborn for visualization:
plt.figure(figsize=(10,4), dpi=200)
sns.countplot(x='division', data=df)
```

```python
df['level of education'].value_counts()
```

```python
# Now we can add an extra category to the plot using the hue argument:
plt.figure(figsize=(10,4), dpi=200)
sns.countplot(x='level of education', data=df, hue='division', palette='Set2')
```

### Barplots:
You must add an extra argument called estimator in which you especify the numpy function that you want to pass.

```python
plt.figure(figsize=(10,4))
sns.barplot(x='level of education', y='salary', data=df, estimator=np.mean)
```

```python
# You can also add a ci argument that stands for the confidence interval:
sns.barplot(x='level of education', y='salary', data=df, estimator=np.mean, ci='sd') # sd stands for standard deviation
```
