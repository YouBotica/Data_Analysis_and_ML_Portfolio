## Seaborn

- Seaborn is a statistical plotting library for python that is specifically designed to interact well with Pandaas Dataframes

- Seaborn is build directly of Matplotlib, but uses simpler "one-line" syntax.

- When useing seaborn, we trade-off customization for ease of use, however, plt commands can still be used for further customization.

## 1. Scatterplots

- Scatter plots show the relationship between two continuous features


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/home/legacy/anaconda3/Data Science Master Class/UNZIP_FOR_NOTEBOOKS_FINAL/05-Seaborn/dm_office_sales.csv')
```

```python
df.head()
```

```python
fig = plt.figure(figsize=(12,8), dpi=200) # You can use matplotlib commands to edit your plot!

sns.scatterplot(x='salary', y='sales', data=df)
```

```python
# I can also make the scatter plot by categories:
fig = plt.figure(figsize=(12,8), dpi=300)
sns.scatterplot(x='salary', y='sales', data=df, hue='level of education')
```

```python
# With hue and marker for classification:
plt.figure(figsize=(12,4), dpi=200)
sns.scatterplot(x='salary', y='sales', data=df, s=200, style='level of education'
                , hue='level of education')

# In order to save the figure:
plt.savefig("FirstScatter.eps")
```
