## Matrix Plots:
- Are the equivalent of displaying graphically a pivot table.

The two main types of matrix plots are:
    1. heatmap()
        - Visually displays the distribution of cell values with a color mapping.
    2. clustermap()
        - Same visual as heatmap, but first conducts hierarchical cllustering to reorganize data into groups.

Note that a heatmap should ideally have all cells be in the same units, so the color mapping makes sense across the entire
dataframe



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/home/legacy/anaconda3/Data Science Master Class/UNZIP_FOR_NOTEBOOKS_FINAL/05-Seaborn/country_table.csv')
```

```python
df = df.set_index("Countries")
```

```python
df.head()
```

```python
sns.heatmap(df.drop('Life expectancy', axis=1), linewidths=0.5, annot=True)
```

```python
# Now, let's see the clustermap:
sns.clustermap(df.drop('Life expectancy', axis=1), linewidths=0.5, annot=True)
```
