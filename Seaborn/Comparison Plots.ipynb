## Comparison plots:
- These are essentially 2-dimensional versions of the plots we've seen so far. 
- The two main plots types discussed here: 
    - jointplot()
    - pairplot()

1. Jointplot:
    - We can map histograms to each feature of a scatterplot to clarify the distributions within each feature.
    - We can also adjust the scatterplot to be a hex plot or a 2D KDE plot.
    - If you do it with hexagons (hex), the darker the hexagon is, the more points are going to be in that area.
    - It is easy to see the trends in XY planes.
    - Also you can use KDEs.

2. Pairplot:
    - It is a quick way to compare all the numerical columns in a dataframe.
    - It automatically creates a histogram for each column and a scatterplot comparison between all possible combinations of columns.
    - Warning: It can be CPU and RAM intensive for huge datasets. Hence, it is a good idea to filter down the data that you are interested in.
    

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
df = pd.read_csv('/home/legacy/anaconda3/Data Science Master Class/UNZIP_FOR_NOTEBOOKS_FINAL/05-Seaborn/StudentsPerformance.csv')
```

```python
df.head()
```

```python
sns.jointplot(data=df, x='math score', y='reading score')
```

```python

# You can use the kind argument to change the plot:
sns.jointplot(data=df, x='math score', y='reading score', kind='hex')
```

```python
sns.pairplot(data=df, hue='gender')
```

```python
# Getting rid of the duplicates:
sns.pairplot(data=df, hue='gender', corner=True)
```
