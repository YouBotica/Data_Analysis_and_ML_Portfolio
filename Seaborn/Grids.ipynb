## Seaborn Grids:

- Seaborn grid calls use matplotlib subplots() to automatically create a grid based off a categorical column.
- Instead of passing in an specific number of cols or rows for the subplots, we can simply pass in the name of the column and
seaborn will automatically map the subplots grid.
- Many of seaborn's calls are been run on top of this grid system.



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/home/legacy/anaconda3/Data Science Master Class/UNZIP_FOR_NOTEBOOKS_FINAL/05-Seaborn/StudentsPerformance.csv')
```

```python
df.head()
```

```python
sns.catplot(data=df, x='gender', y='math score', kind='box', row='lunch')
```

```python
# Now columns:
sns.catplot(data=df, x='gender', y='math score', kind='box', col='lunch')
```

```python
# Now columns:
sns.catplot(data=df, x='gender', y='math score', kind='box', col='lunch', row='test preparation course')
```

Just checking the grid:

```python
g = sns.PairGrid(df) # Empty grid object

# Now let's plot on the empty grid object:
g = g.map_upper(sns.scatterplot)

g = g.map_lower(sns.kdeplot)

g = g.map_diag(sns.histplot)
```
