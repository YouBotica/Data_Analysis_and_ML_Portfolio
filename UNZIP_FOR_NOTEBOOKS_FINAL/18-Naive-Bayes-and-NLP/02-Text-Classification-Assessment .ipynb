### Training and Data

**TASK: Split the data into features and a label (X and y) and then preform a train/test split. You may use whatever settings you like. To compare your results to the solution notebook, use `test_size=0.20, random_state=101`**

```python
#CODE HERE
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(stop_words='english')
```

```python
X_train_tf = tfidf.fit_transform(X_train) # Avoid Data Leakage 
```

```python
X_test_tf = tfidf.transform(X_test) 
```

### Logistic Regression:

```python
from sklearn.metrics import plot_confusion_matrix, classification_report
```

```python
def report(model):
    predictions = model.predict(X_test_tf)
    print(classification_report(y_test, predictions))
    plot_confusion_matrix(model, X_test_tf, y_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
```

```python
log_model = LogisticRegression()
```

```python
param_grid = {'penalty': ['l1', 'l2', 'elasticnet'], # Lasso, Ridge, and both
             'C': [0.01, 0.1, 1.0, 50, 100, 1000],
              'solver': ['lbfgs', 'sag', 'saga']
             }
```

```python
grid_log_model = GridSearchCV(log_model, param_grid=param_grid, cv=10) # Cross Validation
```

```python
grid_log_model.fit(X_train_tf, y_train)
```

```python
grid_log_model.best_params_
```

```python
report(grid_log_model)
```
