## Useful Methods

```python
import pandas as pd
import numpy as np

# Now, let's read an actual CSV file:

df = pd.read_csv('/home/legacy/anaconda3/Data Science Master Class/UNZIP_FOR_NOTEBOOKS_FINAL/03-Pandas/tips.csv')
df
```

```python
# Get the last four digits from the credit card number:
def last_four(num):
    return int(str(num)[-4:])

# Use apply to use that function in the specified collumn:
df['last_four'] = df['CC Number'].apply(last_four) #Don't need to pass an argument for the function
```

```python
last_four
```

## Function that returns $, $$ or $$$ depending on the price of the product

```python
def yelp(price):
    if price < 10:
        return '$'
    elif price >= 10 and price < 30:
        return '$$'
    else:
        return '$$$'
    
df['yelp'] = df['total_bill'].apply(yelp)
df
```

## Quality of the tip: function depending on multiple collumns

```python
def quality(total_bill, tip):
    if tip/total_bill > 0.25:
        return "Generous"
    else:
        return "Other"
```

```python
df['Quality'] = df[['total_bill', 'tip']].apply(lambda df: quality(df['total_bill'], df['tip']), axis=1)
```

```python
# However, there is a much simpler and more effective way to compute this:
df['Quality'] = np.vectorize(quality)(df['total_bill'], df['tip'])

# Vectorize is used when a function is not aware that we are going to pass a Numpy Array
```
