## Sub plots:
-The plt.subplots() method allows us to easily create Figure and Axes objects in side by side formations.
-The plt.subplots() command returns a tuple containing the Figure canvas and then a numpy array holding the axes objects.


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.linspace(0, 10, 11)
b = a ** 4
```

```python
x = np.arange(0,10)
y = 2*x
```

```python
fig,axes = plt.subplots(nrows=2, ncols=2)
axes[0][0].plot(x,y)
axes[1][1].plot(a,b)

plt.tight_layout() #  Makes sure that the graphs are not overlapping
```

```python
# Looping to plot:
fig,axes = plt.subplots(nrows=3,ncols=1)

for ax in axes:
    ax.plot(x,y)

plt.tight_layout() # Makes sure that the graphs are not overlapping
```

```python
fig,axes = plt.subplots(nrows=2, ncols=2)

axes[0][0].plot(x,y)

axes[0][1].plot(a,b)

axes[1][0].plot(a,b)

axes[1][1].plot(x,y)

fig.subplots_adjust(wspace=0.9, hspace=0.4) # wspace and hspace are based on the average axes width and height respectively.

fig.suptitle("Figure Level", fontsize=16)

plt.tight_layout()

fig.savefig('new_subplots.eps', bbox_inches='tight')
```

## Legends:

-Matplotlib offers very robust styling functions that allow us to edit colors, legends, line widths, markers, and much more!

-Note: Due to the wide amount of possible variations, we will use the previous notebooks.

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

x = np.linspace(0,11,10)

ax.plot(x, x, label='X vs X')
ax.plot(x,x**2, label='X vs X^2')
ax.legend() # You can use loc argument to set the location where you want the legend to be displayed

ax.legend(loc=[1.1, 0.5])
```

## Styling:

```python
x = np.linspace(0, 11, 10)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x, color='#e3ba14', label='x vs x', lw=2, ls='--') # RGB Hex code

ax.plot(x,x+1, color='purple', label='x vs x+1', ls='steps')

# Creating your custom markers:
ax.plot(x,x+2, color='red', label='x vs x+2')

ax.legend()
```

```python
lines = ax.plot(x, x+1, color='purple', linewidth=5)
lines[0].set_dashes([5,2,10,2])
```
