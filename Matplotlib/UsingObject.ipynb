## Using the objects for plotting

```python
import matplotlib.pyplot as plt
import numpy as np

# The empty figure object is:
fig = plt.figure() # We  added a size specification of 10 by 10 inches
```

```python
# Adding axes:
axes = fig.add_axes([0, 0, 1, 1]) #x, y, width and height x, y is the origin point of the axes
```

```python
# Let's use these methods for a real function:
a = np.linspace(0, 10, 11)
```

```python
b = a**4
```

```python
x = np.arange(0, 10)
y = 2*x
```

```python
# Create the empty figure object:
fig = plt.figure()
axes1 = fig.add_axes([0, 0, 1, 1])
axes1.plot(a,b)

axes1.set_xlim(0,8)
axes1.set_ylim(0,8000)
axes1.set_xlabel('A')
axes1.set_ylabel('B')
axes1.set_title('Power of 4')

# NOTE: All these commands must be on the same cell.
# Zoomed in plot:

# Small axes:
axes2 = fig.add_axes([0.2, 0.5, 0.25, 0.25])
axes2.plot(a,b)

axes2.set_xlim(1,2)
axes2.set_ylim(0,25)
axes2.set_xlabel('A')
axes2.set_ylabel('B')
axes2.set_title('Zoomed In')
```

## Figure Parameters:

```python
fig = plt.figure(figsize=(10,3), dpi=200)

axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)
# Use bbox_inches parameter in tight mode to be able to visualize the axes when you save the image.
fig.savefig('new_figure.png', bbox_inches='tight')
```
