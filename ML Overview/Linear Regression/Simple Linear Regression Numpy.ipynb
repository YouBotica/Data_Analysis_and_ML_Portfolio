## Simple Linear Regression:


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/home/legacy/anaconda3/Data Science Master Class/UNZIP_FOR_NOTEBOOKS_FINAL/08-Linear-Regression-Models/Advertising.csv')
```

```python
df.info()
```

```python
df.head()
```

Let's combine the sum of the money spent on TV, radio and newspaper to have a single X column for the linear regression:

```python
df['total_spend'] = df['TV'] + df['radio'] + df['newspaper'] + df['sales']
```

```python
df.head()
```

```python
plt.figure(figsize=(12,8), dpi=200)
sns.scatterplot(data=df, x='total_spend', y='sales', s=100)
```

```python
# Seaborn can perform a linear regression using regplot method:
plt.figure(figsize=(12,8), dpi=200)
sns.regplot(data=df, x='total_spend', y='sales')
```

```python
X = df['total_spend']
y = df['sales']
```

```python
betas = np.polyfit(X,y,deg=1)
```

```python
potential_spend  = np.linspace(0, 500, 100)
predicted_sales = betas[0]*potential_spend + betas[-1]
```

```python
sns.scatterplot(data=df, x='total_spend', y='sales')
plt.plot(potential_spend, predicted_sales, color='red')
```

Predicting a value:

```python
spend = 200
predicted_sales = betas[0]*spend + betas[-1]
predicted_sales
```

The problem with this is that we are lacking an overall performance metric. This is the reason why we will be using scikit learn in a future class. For now, let's explore the higher order polynomial fit that numpy offers:

```python
third_order = np.polyfit(X,y,3)
third_order

eq_3 = third_order[0]*potential_spend**3 + third_order[1]*potential_spend**2 + third_order[2]*potential_spend + third_order[-1]
```

```python
sns.scatterplot(data=df, x='total_spend', y='sales')
plt.plot(potential_spend, eq_3)
```
