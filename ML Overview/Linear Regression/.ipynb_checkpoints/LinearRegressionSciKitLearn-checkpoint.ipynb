```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/home/legacy/anaconda3/Data Science Master Class/UNZIP_FOR_NOTEBOOKS_FINAL/08-Linear-Regression-Models/Advertising.csv')
```

```python
df.head()
```

```python
sns.pairplot(data=df)
```

```python
X = df.drop('sales', axis=1)
X
```

```python
y = df['sales']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# For the first time, use the default hyperparameters for training a model, then start playing with these to get
# better results.

model = LinearRegression() # Empty model object
```

```python
model.fit(X_train, y_train) # Fitting
```

```python
# Predicting for all the X_test:
model.predict(X_test)
```

```python
predictions = pd.DataFrame(data=y_test)
```

```python
predictions['y_hat'] = model.predict(X_test)
```

```python
predictions
```

## Evaluating Regression:

Let's discuss the most common evaluation metrics for regression:
1. Mean Absolute Error
2. Mean Squared Error
3. Root Mean Squared Error

These apply to any type of regression problem.



```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
plt.figure(figsize=(10,7), dpi=200)
sns.histplot(data=df, x='sales', bins=20)
```

```python
# MSE
mean_absolute_error(y_test, predictions['y_hat'])
```

```python
# RMSE
np.sqrt(mean_squared_error(y_test, predictions['y_hat']))
```

## Residual plots:

A residual plot will allow me to plot the error, so I don't have to worry about many dimensions.
The residual errors should be random and close to a normal distribution.
- We want the mean of the distribution to be on zero, or very close to zero.


```python
test_residuals = y_test - predictions['y_hat']
test_residuals
```

```python
sns.scatterplot(x=y_test, y=test_residuals)
plt.axhline(y=0, color='red') # Draw a horizontal line

# The residual plot should look random, no clear curve on it.
```

```python
sns.distplot(test_residuals, kde=True, bins=20)

# Residual
```

```python
import scipy as sp
```

```python
fig, ax = plt.subplots(figsize=(6,8), dpi=200)
_ = sp.stats.probplot(test_residuals, plot=ax)
```

```python
final_model = LinearRegression() # Now that we have tested the model, we re-train the model with the complete dataset
# And the same hyperparameters.
```

```python
final_model.fit(X,y)
```

If I want to check de beta coefficients of my model:

```python
final_model.coef_
```

```python
X.head()
```

```python
# We can do a final_model y predictions:
y_hat = final_model.predict(X)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(16,6))

axes[0].plot(df['TV'], df['sales'], 'o', color='blue')
axes[0].plot(df['TV'], y_hat, 'o', color='red')
axes[0].set_ylabel('Sales')
axes[0].set_title('TV Spend')

axes[1].plot(df['radio'], df['sales'], 'o', color='blue')
axes[1].plot(df['radio'], y_hat, 'o', color='red')
axes[1].set_ylabel('Sales')
axes[1].set_title('Radio Spend')

axes[2].plot(df['newspaper'], df['sales'], 'o', color='blue')
axes[2].plot(df['newspaper'], y_hat, 'o', color='red')
axes[2].set_ylabel('Sales')
axes[2].set_title('Newspaper Spend')
```
